```python
import numpy as np
from CharToIndex import CharToIndex
```

```python
f_tegaki = "/net/nfs2/export/home/ohno/CR_pytorch/data/tegaki_katsuji/tegaki.npy"
f_katsuji = "/net/nfs2/export/home/ohno/CR_pytorch/data/tegaki_katsuji/katsuji.npy"

tegaki = np.load(f_tegaki,allow_pickle=True)
katsuji = np.load(f_katsuji,allow_pickle=True)

chars_file_path = "/net/nfs2/export/home/ohno/CR_pytorch/data/tegaki_katsuji/all_chars_3812.npy"
tokens = CharToIndex(chars_file_path)
```

```python
tegaki_inp = tegaki[0]
tegaki_tar = tegaki[1]
tegaki_first_choice = []

for t_inp in tegaki_inp:
    tegaki_first_choice.append(t_inp[0])
```

```python
tegaki_inp_idx = []
tegaki_tar_idx = []

for inp,tar in zip(tegaki_first_choice,tegaki_tar):
    tegaki_inp_idx.append(tokens.get_index(inp))
    tegaki_tar_idx.append(tokens.get_index(tar))
```

```python
collect_char_cnt = np.zeros((len(tokens),2),dtype=int)
wrong_char_cnt   = np.zeros((len(tokens),2),dtype=int)
for inp,tar in zip(tegaki_inp_idx,tegaki_tar_idx):
    collect_char_cnt[tar][1] += 1
    wrong_char_cnt[tar][1] += 1

    if inp==tar:
        collect_char_cnt[tar][0] += 1
    else:
        wrong_char_cnt[tar][0] += 1
```

```python
#正解した回数の多い文字を降順に並べたindices
sorted_collect_char_indices = np.argsort(collect_char_cnt,axis=0)[::-1][:,0]

for idx in sorted_collect_char_indices:
    print(tokens.get_decoded_char(idx),end=': ')
    print(f'{collect_char_cnt[idx][0]}/{collect_char_cnt[idx][1]}回 ({collect_char_cnt[idx][0]/collect_char_cnt[idx][1]:.5}%)')
```

```python
#正解した回数の多い文字を降順に並べたindices
sorted_wrong_char_indices = np.argsort(wrong_char_cnt,axis=0)[::-1][:,0]

for idx in sorted_wrong_char_indices:
    print(tokens.get_decoded_char(idx),end=': ')
    print(f'{wrong_char_cnt[idx][0]}/{wrong_char_cnt[idx][1]}回 ({wrong_char_cnt[idx][0]/wrong_char_cnt[idx][1]:.5}%)')
```
